Question arises ..How the ChatGPT is able to search and give the results instantly?

This search is done by using the below algorithm

 1. **"Approximate Nearest Neighbor"** -- i.Cosine Similarity

                                          ii.Euclidean Distance
                            methods are used to find the nearest match.
  

  i. Cosine Similarity--It is given by the formula Sc(X,Y) =X.Y/||X||x||Y||

  Example:X=(3,2,0,5)
          Y=(1,0,0,0)

          X.Y/||X||x||Y|| =3*1 +2*0 +0*0 + 5*0/SQRT(3**2+2**2 +5**2)SQRT(1**2)
                          =3/SQRT(9+4+25)
                          =0.49

            This shows the two vectors are 49% equal.

  

  ii.Euclidean Distance -- Here two vectors which are at point (x1,y1) and (x2,y2)..we find teh Euclidean distance by using the formula

      dist1= SQRT((x1-x2)**2 + (y1-y2)**2)....distn=SQRT((x1-x2)**2 +(y1-y2)**2)

      Whichever is the nearest it is considered.
