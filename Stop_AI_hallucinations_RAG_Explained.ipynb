**RAG is the process of optimizing the Output of LLM's**

**RAG has two steps involved **
1. Ingestion/Indexing

2. Retreival


1. Indexing --


 a.Consider an internal PDF =====> Break this PDF into multiple chunks ====> Each of these chunks contains textual data. This data is data is converted into Vectors.

 b.These vectors are stored in the "Vector Database. This database is used to store the vector embeddings of the original PDF.
 Example of a Vector DB is Quadrant DB(This DB is quite Fast) .


 The whole Processof breaking the PDF into chunksand converting chunks into Vectors and storing these Vectors in a Database is called as **Indexing orIngestion**

**Vector Database:**

 The Vector database has the concepts of:
 i.Points -- Each chunk size is of this specified "Points"
             "Points" -1241--Each chunk is of size 1241
          

 ii.Segments ---For Performance enchancing this is done..Simulatenoulsy 8 chunks are done.

 iii.Shards---This depends on theFile size or the Content... If the File size is big it is divided into 2 or 3 shards..otherwise it is 1.This is to improve the performance of File loading.


 ** Note** : Both shaards and Segments are used for Scaling purpose. To make the process Faster.


 Cosine Similarity: the Query search is based on **Cosine Similarity**

 **It is a Score based on how similar two items are**

 Example: doc d1,doc d2 are two documents--Check the similarity of the two documents using Cosine similarity..

 d1.d2=d1.d2/||d1||.||d2||

**Numeric Example:**

 Consider two documents X and Y with vectors ith values X(3,2,0,5) and
 Y(1,0,0,0)

 X.Y =3*1+2*0+0*0+0*5 =3

 ||X||=SQRT(3**2 +2**2+0**2+5**2) =6.16

 ||Y|| =SQRT(1**2+0**2+0**2+0**2)=1

 Similarity Score (Sc) =X.Y/||X||.||Y|| =0.49

 Dis Similarity Score (Dc) = 1-0.49 = 0.51
